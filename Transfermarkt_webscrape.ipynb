```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.transfermarkt.com/liga-portugal/transfers/wettbewerb/PO1/plus/?saison_id=2018&s_w=&leihe=0&leihe=1&intern=0&intern=1'

response = requests.get(url, headers={'User-Agent': 'Custom5'})
print(response.status_code)
finance_data = response.text
soup = BeautifulSoup(finance_data, 'html.parser')
```

```python
years_url = list(range(2004, 2022))
# último número da lista (2022) vai ser excluído
```

**Webscrape dos valores de venda de jogadores**

```python
income_list = []
income_per_club_list = []
income_per_player_list = []
year_list = []


for year_url in years_url:

    url = 'https://www.transfermarkt.com/liga-portugal/transfers/wettbewerb/PO1/plus/?saison_id=' + str(year_url) + '&s_w=&leihe=0&leihe=1&intern=0&intern=1'

    response = requests.get(url, headers={'User-Agent': 'Custom5'})
    # print(response.status_code)
    finance_data = response.text
    soup = BeautifulSoup(finance_data, 'html.parser')

    grouped_data = soup.find('div', {'class': 'transferbilanz'})

    income = int(grouped_data.find_all('span', {'class': 'greentext'})[0].text.replace('€','').replace(',',''))
    income_per_club = int(grouped_data.find_all('span', {'class': 'greentext'})[1].text.replace('€','').replace(',', ''))
    income_per_player = int(grouped_data.find_all('span', {'class': 'greentext'})[2].text.replace('€','').replace(',', ''))
    year = year_url   

    income_list.append(income)
    income_per_club_list.append(income_per_club)
    income_per_player_list.append(income_per_player)
  
    year_list.append(year)
```

**Webscrape dos valores de venda de jogadores**

```python
Transfer_expenses_list = []
Expenditures_per_club_list = []
Expenditures_per_player_list = []
year_list = []


for year_url in years_url:

    url = 'https://www.transfermarkt.com/liga-portugal/transfers/wettbewerb/PO1/plus/?saison_id=' + str(year_url) + '&s_w=&leihe=0&leihe=1&intern=0&intern=1'

    response = requests.get(url, headers={'User-Agent': 'Custom5'})
    # print(response.status_code)
    finance_data1 = response.text
    soup = BeautifulSoup(finance_data1, 'html.parser')

    grouped_data = soup.find('div', {'class': 'transferbilanz'})

    Transfer_expenses = int(grouped_data.find_all('span', {'class': 'redtext'})[0].text.replace('€','').replace(',',''))
    Expenditures_per_club = int(grouped_data.find_all('span', {'class': 'redtext'})[1].text.replace('€','').replace(',', ''))
    Expenditures_per_player = int(grouped_data.find_all('span', {'class': 'redtext'})[2].text.replace('€','').replace(',', ''))
    year = year_url   

    Transfer_expenses_list.append(Transfer_expenses)
    Expenditures_per_club_list.append(Expenditures_per_club)
    Expenditures_per_player_list.append(Expenditures_per_player)
  
    year_list.append(year)
```

```python
ligapt = pd.DataFrame({'Year': year_list,
                           'Income': income_list,
                           'Income_per_club': income_per_club_list,
                           'Income_per_player': income_per_player_list,
                           'Transfer_expenses': Transfer_expenses_list,
                           'Expenditures_per_club': Expenditures_per_club_list,
                           'Expenditures_per_player': Expenditures_per_player_list
                            })
```

**Cálculo do balanço do mercado de transferências**

```python
ligapt['Total_Balance']=ligapt['Income']-ligapt['Transfer_expenses']
ligapt['Balance_per_club']=ligapt['Income_per_club']-ligapt['Expenditures_per_club']
ligapt['Balance_per_player']=ligapt['Income_per_player']-ligapt['Expenditures_per_player']

ligapt
```

```python
ligapt.to_excel('rendas_transf_pt.xlsx', index=False)
```
