```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import unidecode

pd.set_option('display.max_columns', None)
```

# SCRAPING PROCESS (THERE IS NO NEED TO RUN THIS, THIS CODE IS ONLY FOR DEBUGGING)

### Extracting team urls

```python
url = "https://www.bundesliga.com/en/bundesliga/table"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

teams = soup.find_all('td', class_='team')

umlaut_mapping = {
    'ä': 'ae',
    'ö': 'oe',
    'ü': 'ue',
    'Ä': 'Ae',
    'Ö': 'Oe',
    'Ü': 'Ue'
}

team_urls = []
for team in teams:
    div_tag = team.find('div')
    if div_tag and 'title' in div_tag.attrs:
        team_name = div_tag['title']

        # Replace Umlauts
        for umlaut, replacement in umlaut_mapping.items():
            team_name = team_name.replace(umlaut, replacement)
        # Remove ponctuation
        team_name = unidecode.unidecode(team_name)
        # Replace spaces by hifens, remove dots convert to lowercase
        team_name = team_name.replace(' ', '-').replace('.', '').lower()
        team_url = "https://www.bundesliga.com/en/bundesliga/clubs/" + team_name + "#squad"
        team_urls.append(team_url)
```

```python
team_urls
```

###  Now that we every team url, let's iterate for each url and get every player url

```python
all_player_names = []

for team_url in team_urls:
    response = requests.get(team_url)
    soup = BeautifulSoup(response.content, "html.parser")

    player_cards = soup.find_all("div", class_="playercard-wrapper")

    for player_card in player_cards:
        player_image = player_card.find("img")
        if player_image and "alt" in player_image.attrs:
            alt_value = player_image["alt"]
            alt_value = unidecode.unidecode(alt_value)
            alt_value = alt_value.replace(' ', '-').lower()
            all_player_names.append('https://www.bundesliga.com/en/bundesliga/player/' + alt_value)
```

```python
def scrape_german_football_data(league="bundesliga"):

  url = f"https://www.bundesliga.com/en/{league}/table"

  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')

  teams = soup.find_all('td', class_='team')

  umlaut_mapping = {
      'ä': 'ae',
      'ö': 'oe',
      'ü': 'ue',
      'Ä': 'Ae',
      'Ö': 'Oe',
      'Ü': 'Ue'
  }

  team_urls = []
  for team in teams:
      div_tag = team.find('div')
      if div_tag and 'title' in div_tag.attrs:
          team_name = div_tag['title']

          # Replace Umlauts
          for umlaut, replacement in umlaut_mapping.items():
              team_name = team_name.replace(umlaut, replacement)
          # Remove ponctuation
          team_name = unidecode.unidecode(team_name)
          # Replace spaces by hifens, remove dots convert to lowercase
          team_name = team_name.replace(' ', '-').replace('.', '').lower()
          team_url = f"https://www.bundesliga.com/en/{league}/clubs/{team_name}#squad"
          team_urls.append(team_url)



  all_player_names = []

  for team_url in team_urls:
      response = requests.get(team_url)
      soup = BeautifulSoup(response.content, "html.parser")

      player_cards = soup.find_all("div", class_="playercard-wrapper")

      for player_card in player_cards:
          player_image = player_card.find("img")
          if player_image and "alt" in player_image.attrs:
              alt_value = player_image["alt"]
              alt_value = unidecode.unidecode(alt_value)
              alt_value = alt_value.replace(' ', '-').lower()
              all_player_names.append(f'https://www.bundesliga.com/en/{league}/player/{alt_value}')



  columns = ["Player", "Position", "Birthday", "Nationality", "Height", "Weight", "Goals", "Assists", "Shots", "Shots against post and bar", "Penalties", "Penalties scored", "Tackles won", "Aerial duels won", "Fouls", "Yellow cards", "Appearances", "Sprints", "Intensive runs", "Distance (km)", "Speed (km/h)", "Passes completed", "Own-goals", "Crosses", "Appearances", "Goals", "Assists", "Ball actions", "Distance (km)", "Penalties"]

  player_data = []

  for player_url in all_player_names:
      response = requests.get(player_url)
      soup = BeautifulSoup(response.content, "html.parser")

      info_elements = soup.find_all("div", class_="info col-12 col-md-6")
      info_elements_x = soup.find_all("div", class_="info col-12 col-md-6 ng-star-inserted")

      position_element = next((elem for elem in info_elements if "Position" in elem.text), None)
      birthday_element = next((elem for elem in info_elements if "Birthday" in elem.text), None)
      nationality_element = next((elem for elem in info_elements if "Nationality" in elem.text), None)

      height_element = next((elem for elem in info_elements_x if "Height" in elem.text), None)
      weight_element = next((elem for elem in info_elements_x if "Weight" in elem.text), None)

      try:
          position = position_element.find("span", class_="value").text.strip() if position_element else None
          birthday = birthday_element.find("span", class_="value").text.strip() if birthday_element else None
          nationality = nationality_element.find("span", class_="value").text.strip() if nationality_element else None
      except AttributeError:
          position, birthday, nationality = None, None, None

      height = height_element.find("span", class_="value").text.strip() if height_element else None
      weight = weight_element.find("span", class_="value").text.strip() if weight_element else None

      #print(player_url)

      if position != "Goalkeeper":
          stat_values = soup.find_all("div", class_="col col-4 value")
          values = [float(element.text.strip()) for element in stat_values]

          stat_cols = soup.find_all("div", class_="col col-8 key")
          col_values = [element.text.strip() for element in stat_cols]

          player_elements = soup.find_all("div", class_="col-8")
          player_name = player_elements[0].text.strip().split("of")[1].strip() if player_elements else None

          columns = ["Player", "Position", "Birthday", "Nationality", "Height", "Weight"] + col_values
          data = [player_name, position, birthday, nationality, height, weight] + values

          player_data.append(data)

  df = pd.DataFrame(player_data, columns=columns)

  return df
```
