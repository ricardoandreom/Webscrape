```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
import numpy as np
```

```python
product = 'relogio'
num_pages_to_scrape = 5
base_url = 'https://www.amazon.es/-/pt/s?k=' + product + '&page='
headers = {'User-Agent': 'Chrome/58.0.3029.110'}

df_final = pd.DataFrame()

for page_num in range(2, num_pages_to_scrape + 1):
    url = base_url + str(page_num)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    products = soup.find_all('div', {'class': 'sg-col-inner'})

    descriptions = []
    brands = []
    prices = []
    ratings = []
    delivery_infos = []

    product_data_format =[]

    for product in products:

        description_element = product.find('h2', {'class': 'a-size-mini a-spacing-none a-color-base s-line-clamp-2'})
        brand_element = product.find('h2', {'class': 'a-size-mini s-line-clamp-1'})
        price_element = product.find('span', {'class': 'a-price'})
        rating_element = product.find('i', {'class': 'a-icon a-icon-star-small a-star-small-4-5 aok-align-bottom'})
        delivery_element = product.find('div', {'class': 'a-row a-size-base a-color-secondary s-align-children-center'})

        brand = brand_element.get_text().strip() if brand_element else np.nan
        description = description_element.get_text().strip() if description_element else np.nan
        price = price_element.get_text().strip() if price_element else np.nan
        rating = rating_element.get_text().strip() if rating_element else np.nan
        delivery_info = delivery_element.get_text().strip() if delivery_element else np.nan

        # Store these extracted values
        descriptions.append(description)
        brands.append(brand)
        prices.append(price)
        ratings.append(rating)
        delivery_infos.append(delivery_info)

    data = {
        'Description': descriptions,
        'Brand': brands,
        'Price': prices,
        'Rating': ratings,
        'Delivery Info': delivery_infos
         }

    df = pd.DataFrame(data)
    df_final = pd.concat([df_final, df])

```

```python
def clean_process_data(df, product):
  df['Product'] =  product
  nan_count = df.isnull().sum(axis=1)
  df = df[nan_count <= 3]
  df['Rating'] = df['Rating'].str.extract(r'(\d+\.\d+)', expand=False)
  df['Price'] = df['Price'].str.extract(r'(\d+,\d+)', expand=False)
  df['Delivery_is_free'] = df['Delivery Info'].str.contains('GRÁTIS', case=False, na=False)
  df['Delivery_is_free'] = df['Delivery_is_free'].map({True: 'Yes'}).fillna('No')
  df = df[['Product', 'Brand', 'Description', 'Price', 'Rating', 'Delivery Info', 'Delivery_is_free']]

  return df
```

```python
df = clean_process_data(df_final, 'Relogio')
```
